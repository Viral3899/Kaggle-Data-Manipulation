```python
import pandas as pd
pd.set_option("display.max_rows", 10)
df=pd.read_csv('winemag-data-130k-v2.csv',index_col=0)
df
```

```python
df.groupby('points')['points'].count()
```

```python
df.groupby('points')['price'].min()
```

```python
df.groupby('winery').apply(lambda df1: df1.title.iloc[0])
```

```python
df.groupby(['country', 'province']).apply(lambda df1 : df1.loc[df1.points.idxmax()])
```

```python
df.groupby(['country'])['price'].agg([len, min, max])
```

```python
countries_reviewed = df.groupby(['country', 'province']
                               )['description'].agg([len])
countries_reviewed
```

### Practice

```python
reviews_written = df.groupby('taster_twitter_handle')['taster_twitter_handle'].count()
print(reviews_written)
```

```python
best_rating_per_price =  df.groupby('price')['points'].max().sort_index()
print(best_rating_per_price)
```

```python
price_extremes = df.groupby(['variety'])['price'].agg(['min','max'])
print(price_extremes)
```

```python
sorted_varieties = price_extremes.sort_values(by=['min','max'],ascending=False)
print(sorted_varieties)
```

```python
reviewer_mean_ratings = df.groupby(['taster_name'])['points'].mean()
reviewer_mean_ratings
```

```python
reviewer_mean_ratings.describe()
```

```python
country_variety_counts = df.groupby(['country','variety']).size().sort_values(ascending=False)
print(country_variety_counts)
```
